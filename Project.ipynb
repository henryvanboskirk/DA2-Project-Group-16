```python
import os
import cv2
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load CSV files
train_df = pd.read_csv('train.csv')

# Define a function to preprocess images
def preprocess_image(file):
    # Check if the file exists
    if not os.path.exists(file):
        print(f'File not found: {file}')
        return None

    img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)

    # Check if the image is correctly loaded
    if img is None:
        print(f'Could not load image: {file}')
        return None

    img = cv2.resize(img, (150, 150))
    img = img / 255.0
    return img.reshape(150, 150, 1)

# Preprocess train images
train_images = np.array([preprocess_image(os.path.join('train_images', f'{img_id}.jpg')) for img_id in train_df['id']])

# Extract target labels
target_columns = ['X4', 'X11', 'X18', 'X26', 'X50', 'X3112']  # replace with your actual target columns
train_labels = train_df[target_columns]

# Split the train data
X_train, X_val, y_train, y_val = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)

# Define the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(len(target_columns), activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val))

# Preprocess test images
test_df = pd.read_csv('test.csv')
test_images = np.array([preprocess_image(os.path.join('test_images', f'{img_id}.jpg')) for img_id in test_df['id']])

# Make predictions on the test data
predictions = model.predict(test_images)
```
